```python
from math import comb
```

$$ f(n,k,p) = P\{X = k\} = \binom{n}{k} p^{k} (1-p)^{n-k} $$

```python
def f(n, k, p):
    return comb(n, k) * p**k * (1 - p)**(n - k)
```

$$ $$

$$ F(n,k,p) = P\{X <= k\} = \sum_{i=0}^{n} \binom{n}{i} p^{i} (1-p)^{n-i} $$

```python
def F(n, k, p):
    return sum(f(n, i, p) for i in range(k + 1))
```

$$ $$

$$ P\{X = k + 1\} = \dfrac{p}{1-p}\dfrac{n - k}{k + 1}P\{X = k\} $$

```python
def F(n, k, p):
    v = [f(n, 0, p)]
    for i in range(1, k + 1):
        v.append((p / (1 - p)) * ((n - (i + 1)) / (i)) * v[i - 1])

    return sum(v)
```

$$ $$

### Example
Assume $X$ is a Binomial random variable with parameters $(n = 6,\space p = 0.4)$.
<br>
Calculate $P\{X = i\}, i = 0..6$


```python
n = 6
p = 0.4

for i in range(n + 1):
    print(f"P{{X = {i}}} = {f(n, i, p):0.4f}")
```

```python
def CDF(n, p):
    x = np.arange(1, n + 1)
    sns.lineplot(x = x, y = np.vectorize(F)(n, x, p))
```
